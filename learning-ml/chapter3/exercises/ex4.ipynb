```python
import utils
import numpy as np
features = np.array([5])
labels = np.array([15])

print(features)
print(labels)
utils.plot_points(features, labels)
```

```python
def absolute_trick(offset, weight, x_point, y_point, learning_rate):
    predicted_y_point = offset + weight*x_point
    if y_point > predicted_y_point:
        weight += learning_rate*x_point
        offset += learning_rate
    else:
        weight -= learning_rate*x_point
        offset -= learning_rate
    return weight, offset
```

```python
def square_trick(offset, weight, x_point, y_point, learning_rate):
    predicted_y_point = offset + weight*x_point
    weight += learning_rate*x_point*(y_point-predicted_y_point)
    offset += learning_rate*(y_point-predicted_y_point)
    return weight, offset
```

```python
weight, offset = 2.0, 3.0

utils.plot_points(features, labels)
utils.draw_line(weight, offset, color='gray', starting=0, ending=8)
for epoch in range(100):
    weight, offset = absolute_trick(weight, offset, 5.0, 15.0, 0.01)
    utils.draw_line(weight, offset, color='lightgray', starting=0, ending=8)
utils.draw_line(weight, offset, color='black', starting=0, ending=8)
```

```python
weight, offset = 2.0, 3.0

utils.plot_points(features, labels)
utils.draw_line(weight, offset, color='gray', starting=0, ending=8)
for epoch in range(5):
    weight, offset = square_trick(weight, offset, 5.0, 15.0, 0.01)
    utils.draw_line(weight, offset, color='lightgray', starting=0, ending=8)
utils.draw_line(weight, offset, color='black', starting=0, ending=8)
```
